```python
Q1 = df['alcohol'].quantile(0.25)
Q3 = df['alcohol'].quantile(0.75)
```

```python
IQR = Q3-Q1
```

```python
lower_fence = Q1 -1.5*IQR
upper_fence = Q3 + 1.5 *IQR
```

```python
lower_fence
```

```python
upper_fence
```

```python
df['alcohol'] = np.where((df.alcohol >= upper_fence) | (df.alcohol <= lower_fence), df['alcohol'].mean() , df['alcohol'])
```
