```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#genratic syntheic data points
```

```python
np.random.seed(1)
```

```python
X = 2 * np.random.rand(100,1)
```

```python
X
```

```python
# genratic Y 
```

```python
y =  4 + 3 * X + 1.5* X**2 + np.random.randn(100,1)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python


X_train , X_test , y_train , y_test = train_test_split(X , y ,test_size= .2 , random_state=42 )
```

```python
X_train
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
degree = 2 
ploy_feature = PolynomialFeatures(degree = 2  , include_bias = False )
```

```python
X_poly_train = ploy_feature.fit_transform(X_train)
```

```python
X_poly_test = ploy_feature.transform(X_test)
```

```python
X_poly_test
```

```python
X_poly_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
poly_reg = LinearRegression()
```

```python
poly_reg
```

```python
poly_reg.fit(X_poly_train , y_train)
```

```python
poly_reg.coef_
```

```python
poly_reg.intercept_
```

```python
y_poly_pred = poly_reg.predict(X_poly_train)
```

```python
y_poly_predss = poly_reg.predict(X_poly_test)
```

```python
y_poly_pred
```

```python
y_poly_predss
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train, y_poly_pred)
```

```python
mean_squared_error(y_test, y_poly_predss)
```

```python
plt.scatter(X_train , y_train )
```

```python
plt.scatter(X_train , y_train)
plt.plot(X_train,y_poly_pred )
plt.show()
```

```python
plt.scatter(X_train, y_train, label = "training data")
X_range = np.linspace(0, 2, 100).reshape(-1, 1)
X_range_poly = ploy_feature.transform(X_range)
plt.plot(X_range, poly_reg.predict(X_range_poly), color='red', label=f'Polynomial Regression (Degree {degree})')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
