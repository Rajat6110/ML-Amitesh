```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
```

```python
df
```

```python
df.drop('name' , axis = 1 , inplace = True)
```

```python
df.isna().sum()
```

```python
median = df['horsepower'].median()
```

```python
median
```

```python
df['horsepower'] = df['horsepower'].fillna(median)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encode = LabelEncoder()
```

```python
encode
```

```python
encode.fit(df['origin'])
```

```python
df['origin'] = encode.transform(df['origin'])
```

```python
df.drop(['origin' , 'cylinders' ,'acceleration'] , axis =1 , inplace = True )
```

```python
df
```

```python
X = df.drop('mpg',axis =1 )
```

```python
y = df['mpg']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=1)
```

```python
#simple linear 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred= model.predict(X_test)
```

```python
for i,col_names in enumerate(X_train.columns):
    print(f"The feature  {col_names} is having coef {model.coef_[i]}")
```

```python
# LAsso
```

```python
from sklearn.metrics import r2_score
```

```python
r2_linear = r2_score(y_test,y_pred)
```

```python
r2_linear
```

```python
y_pred_lasso = lass0.predict(X_test)
```

```python
for i,col_names in enumerate(X_train.columns):
    print(f"The feature  {col_names} is having coef {lass0.coef_[i]}")
```

```python
r2_score(y_test , y_pred_lasso)
```
