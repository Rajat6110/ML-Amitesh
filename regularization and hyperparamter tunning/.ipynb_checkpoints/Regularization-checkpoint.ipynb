# CV WITH HYPERPARAMETER

```python
# Grid Serach CV 
```

```python
from sklearn.model_selection import GridSearchCV , RandomizedSearchCV
```

```python
from sklearn.linear_model import Lasso, Ridge , LinearRegression
```

```python
lasso = Lasso()
```

```python
lasso
```

```python
param_gird = {'alpha': [0.001 , 0.01 , 0.1 , 1 , 10 , 100]}
```

```python
param_gird
```

```python
grid = GridSearchCV(estimator= lasso , param_grid= param_gird , cv = 5 , scoring= 'r2' , verbose=2)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
ridge = Ridge()
```

```python
grid_ridge = GridSearchCV(estimator= ridge , param_grid= param_gird , cv = 5 , scoring= 'r2' , verbose=2)
```

```python
grid_ridge.fit(X_train,y_train)
```

```python
grid.best_params_['alpha']
```

```python
grid.best_score_ # which accuracy was best
```

```python
grid_ridge.best_params_['alpha']
```

```python
grid_ridge.best_score_
```

```python
grid.best_estimator_
```

```python
y_pred_grid_lasso= grid.best_estimator_.predict(X_test)
```

```python
y_pred_grid_lasso_r2 = r2_score(y_test , y_pred_grid_lasso)
```

```python
y_pred_grid_lasso_r2
```

```python
# R2 OF RIDGE WITH GRIDCV
```

```python
grid_ridge.best_estimator_
```

```python
y_pred_ridge_cv = grid_ridge.best_estimator_.predict(X_test)
```

```python
y_pred_grid_ridge_r2 = r2_score(y_test , y_pred_ridge_cv)
```

```python
y_pred_ridge_cv
```

```python
y_pred_grid_ridge_r2
```
