```python
# Pipeline
# Pipeline is a way to chain multiple steps (preprocessing + model) into one object so your whole ML flow becomes a single unit.

# Example flow:

# Scaling → Encoding → Model training


# All run together with:

# fit() → transform() → predict()


# Why it’s useful:

# No data leakage

# Clean, organized code

# Works perfectly with GridSearchCV

# Reproducible workflow
```

```python
# ColumnTransformer
# ColumnTransformer is used when different columns need different preprocessing.

# Real data has:

# Numerical columns → scaling

# Categorical columns → encoding

# ColumnTransformer handles this in one place.
```

```python
from sklearn.pipeline import Pipeline # seqnece of data transformer
```

```python
from sklearn.compose import ColumnTransformer # groups all the piplines steps for each of the columns
```

```python
cat_cols = ["sex" ,"smoker" , "day"]
num_cols = [ 'tip' , 'size']
```

```python
cat_cols
```

```python
num_cols
```

```python
# feATURE ENGG AUTOMATION USING PIPLINE AND COL TRANSFORMER 
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

numerical_pipeline = Pipeline(steps=[
    ("imputation", SimpleImputer(strategy="median")),
    ("scaling", StandardScaler())
])

categorical_pipeline = Pipeline(steps=[
    ("imputation", SimpleImputer(strategy="most_frequent")),
    ("encoding", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
preprocessor = ColumnTransformer(transformers=[
    ("numerical_pipeline", numerical_pipeline, num_cols),
    ("categorical_columns", categorical_pipeline, cat_cols)
])
```

```python
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
for i in range(len(models)):
    print(list(models.values())[i]) # values bcz it is value 
```

```python
from sklearn.metrics import accuracy_score
```

```python
def model_train_eval(X_train , X_test , y_train , y_test , models):
    evaluation = {}
    for i in range(len(models)):
        model = list(models.values())[i]
        model.fit(X_train , y_train)
        y_pred = model.predict(X_test)
        model_score = accuracy_score(y_test , y_pred)
        evaluation[list(models.keys())[i]] = model_score
    return evaluation
```

```python
model_train_eval(X_train , X_test , y_train ,y_test,models)
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
}

results = model_train_eval(X_train, X_test, y_train, y_test, models)
print(results)
```
