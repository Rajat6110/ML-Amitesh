```python
df = pd.DataFrame({'time':['lunch', 'breakfast', 'dinner', 'lunch', 'breakfast', 'dinner', 'lunch', 'breakfast', 'dinner'],
             'total_bill': [120, 130, 90, 125, 150, 190, 160, 180, 189]})
```

```python
df
```

```python
mean_price= df.groupby('time')['total_bill'].mean().to_dict()
```

```python
df['time_encoded'] = df['time'].map(mean_price)
```

```python
df
```

```python
# Label encoding and ordinal encoding are both techniques for converting categorical data into numerical format,
# but they differ in how they handle categorical variables with inherent order. 
# Label encoding assigns a unique integer to each category without considering any relationship between them,
# while ordinal encoding assigns integers based on the order or rank of categories
```
