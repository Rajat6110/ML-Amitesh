```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
sns.distplot(df['total_bill'])
```

```python
mean = np.mean(df['total_bill'])
```

```python
mean
```

```python
std = np.std(df['total_bill'])
```

```python
std
```

```python
std_data = []
for i in list(df['total_bill']):
    zscore = i- mean / std
    std_data.append(zscore)
```

```python
std_data
```

```python
sns.distplot(std_data)
```

```python
# feature scaling is optional , it does'nt chnage the distribution of data 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill']]) # this will calculate mu and sigma 
```

```python
scaler.transform(df[['total_bill']])
```

```python
# scaling muliple cols 
```

```python
scaler.fit_transform(df[['total_bill' , 'tip']])
```

```python
pd.DataFrame(scaler.fit_transform(df[['total_bill' , 'tip']]) , columns = ['bill' , 'tip'])
```

```python
scaler.transform([[13,12]])
```

```python
# normalization / min max scaling 
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
```

```python
min_max.fit_transform(df[['total_bill' , 'tip']])
```

```python
min_max.transform([[10,32]]) # for unseen / new data we only use transfrom
```
