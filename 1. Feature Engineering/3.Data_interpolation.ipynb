```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
#data interpolation >> is the process of estimatinng unknoen valuves from known values 

# types 
#linear 
#cubic 
# polynomial
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,5,7])
```

```python
plt.scatter(x,y)
```

```python
#linear interpolation >> linear in nature 
```

```python
x_new = np.linspace(1,5,10)
y_interp = np.interp(x_new , x,y)
```

```python
plt.scatter(x_new , y_interp)
```

# cucbic interpolation 

```python
x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])
```

```python
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
f = interp1d(x,y, kind = 'cubic')
```

```python
f
```

```python
x_new = np.linspace(1,5,10)
```

```python
y_interp =  f(x_new)
```

```python
plt.scatter(x_new , y_interp)
```

```python
#polynomial interpolation 
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,1,25])
```

```python
plt.scatter(x,y)
```

```python
p = np.polyfit(x,y,2) #2 -> degree 2 hai 
```

```python
x_new = np.linspace(x,y,2)
y_interp = np.polyval(p , x_new)
```

```python
plt.scatter(x_new,y_interp)
```
