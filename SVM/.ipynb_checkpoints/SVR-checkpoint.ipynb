```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_regression
X,y = make_regression(n_samples= 1000 , n_features=2 , n_targets=1 , noise = 3)
```

```python
X
```

```python
y.shape
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0] , y = pd.DataFrame(X)[1] , hue = y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size=0.3 , random_state=1)
```

```python
X_train.shape
```

```python
from sklearn.svm import SVR
```

```python
reg = SVR()
```

```python
reg
```

```python
reg.fit(X_train, y_train)
```

```python
y_Pred = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score , mean_absolute_error
```

```python
r2_score(y_test,y_Pred)
```

# HYPERPARAMETER TUNNING WITH SVR 

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'C': [ 1, 2, 3, 10, 50, 120, 100],
         'gamma': [1, 0.1, 0.2, 0.001, 0.003],
         'kernel': ['linear'],
         'epsilon': [0.01, 0.1, 0.2, 0.3]}
```

```python
g = GridSearchCV(reg , cv = 5 , scoring='r2' , param_grid= params , verbose = 2)
```

```python
g.fit(X_train , y_train)
```

```python
gg = g.best_estimator_
```

```python
yy_pred = gg.predict(X_test)
```

```python
r2_score(y_test , yy_pred)
```

```python
mean_absolute_error(y_test , yy_pred)
```

```python
x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 100),
    np.linspace(y_min, y_max, 100)
)

Z = gg.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X[:,0], X[:,1], c=y, edgecolors='k')
plt.colorbar(label="Predicted y value")
plt.title("SVR Prediction Surface (Best from GridSearchCV)")
plt.show()
```
