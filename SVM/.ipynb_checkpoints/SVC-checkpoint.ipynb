```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000 , n_features=2 , n_classes=2 , n_clusters_per_class=2 , n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)
```

```python
pd.DataFrame(X)[0] # --> access the first column 
```

```python
y
```

```python
sns.scatterplot(x = pd.DataFrame(X)[0] , y = pd.DataFrame(X)[1] , hue = y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size=0.3 , random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC(kernel='linear')
```

```python
clf.fit(X_train, y_train)
```

```python
y_p = clf.predict(X_test)
```

```python
y_p
```

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
accuracy_score(y_test,y_p)
```

```python
print(classification_report(y_test,y_p))
```

# HYPERPARAMETER TUNNING 

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'C': [0.1, 0.2, 1, 2, 3, 10, 50, 100],
         'gamma': [1, 0.1, 0.2, 0.001, 0.003],
         'kernel': ['linear']}
```

```python
g = GridSearchCV(clf , param_grid = params  , cv=5  , scoring='accuracy' , verbose = 3)
```

```python
g.fit(X_train , y_train)
```

```python
g.best_estimator_
```

```python
g.best_index_
```

```python
g.best_params_
```

```python
g.cv
```

```python
g.error_score
```

```python
y_pp = g.best_estimator_.predict(X_test)
```

```python
accuracy_score(y_pp , y_test)
```

```python
x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

Z = g.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:,0], X[:,1], c=y, edgecolors='k')
plt.scatter(clf.support_vectors_[:,0],
            clf.support_vectors_[:,1],
            s=100, facecolors='none', edgecolors='red')

plt.title("SVM Classification Visualization")
plt.show()
```
